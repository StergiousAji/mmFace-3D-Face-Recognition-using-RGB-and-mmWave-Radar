```python
import matplotlib.pyplot as plt
import numpy as np

t = np.sin(np.arange(256))
sp = np.fft.fft(t)
freq = np.fft.fftfreq(t.shape[-1])
# plt.plot(freq, t)
plt.plot(freq, sp.real, freq, sp.imag)
plt.show()
```
